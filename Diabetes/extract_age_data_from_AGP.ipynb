## Étape 4: filtrage des données
Avec les métadonnées, on a une super grosse table BIOM, qui contient toutes les séquences et tous les échantillons (chaque colonne est un échantillon et chaque rangée est une séquence). La table s'appelle AGP.data.biom.qza, et on va la filtrer avec nos sous-ensembles de métadonnées.

```python
#first, filter biom table to only the features in df_t2d
```

```python
conda activate qiime2-metagenome-2024.10
```

```python

qiime feature-table filter-samples --i-table AGP.data.biom.qza --m-metadata-file ../AGP_ages.metadata.txt --o-filtered-table AGP.data.biom.filtered_samples.ages.qza
```

```python
#then, remove all features with low count
```

```python
qiime feature-table filter-features-conditionally --i-table AGP.data.biom.filtered_samples.ages.qza --o-filtered-table AGP.data.biom.filtered_samples.ages.filtered_features.qza --p-abundance 0.001 --p-prevalence 0.001
```

```python
#then, export the table and get the sequences and map to greengenes
```

```python
!qiime tools export --input-path AGP.data.biom.filtered_samples.ages.filtered_features.qza --output-path AGES
!cd FILTERED_T2D_BIOM
!biom convert -i feature-table.biom -o AGP.data.biom.filtered.ages.tsv --to-tsv
!biom table-ids -i feature-table.biom --observations |     awk '{ print ">" $1 "\n" $1 }' > AGP_ages_seqs.fa
```

```python
#now lets do the same thing for our ocntrol dataset
!qiime feature-table filter-samples --i-table AGP.data.biom.qza --m-metadata-file ../AGP_T2Dcontrol.metadata.txt --o-filtered-table AGP.data.biom.filtered_samples.noT2D.qza
```

```python
!qiime feature-table filter-features-conditionally --i-table AGP.data.biom.filtered_samples.noT2D.qza --o-filtered-table AGP.data.biom.filtered_samples.noT2D.filtered_features.qza --p-abundance 0.01 --p-prevalence 0.01
```
